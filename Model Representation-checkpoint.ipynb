```python
# Converting our model architecture in tensorflow keras so that we can have a visual representation of the model

from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Flatten, Conv2D, Dropout, MaxPooling2D, InputLayer, ZeroPadding2D, Activation
from collections import defaultdict
import visualkeras
from PIL import ImageFont
from tensorflow.python.keras.layers import GlobalAveragePooling2D

# create CNN model
image_size = 224
model = Sequential()

# Layer 1
model.add(Conv2D(64, kernel_size=3, strides=1, padding='same', input_shape=(224, 224, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(visualkeras.SpacingDummyLayer())

# Layer 2
model.add(Conv2D(128, kernel_size=3, strides=1, padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(visualkeras.SpacingDummyLayer())

# Layer 3
model.add(Conv2D(256, kernel_size=3, strides=1, padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(visualkeras.SpacingDummyLayer())

# Layer 4
model.add(Conv2D(256, kernel_size=3, strides=1, padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(visualkeras.SpacingDummyLayer())

# Layer 5
model.add(Conv2D(512, kernel_size=3, strides=1, padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(visualkeras.SpacingDummyLayer())

# Layer 6
model.add(Conv2D(512, kernel_size=3, strides=1, padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(visualkeras.SpacingDummyLayer())

# Dropout
model.add(Dropout(0.5))

# Global Average Pooling
model.add(Flatten())

# Dense 1
model.add(Dense(4096))
model.add(Activation('relu'))

# Dense 2
model.add(Dense(4096))
model.add(Activation('relu'))

# Output layer
model.add(Dense(7, activation='softmax'))

# Now visualize the model!

color_map = defaultdict(dict)
color_map[Conv2D]['fill'] = 'orange'
color_map[ZeroPadding2D]['fill'] = 'gray'
color_map[Dropout]['fill'] = 'pink'
color_map[MaxPooling2D]['fill'] = 'red'
color_map[Dense]['fill'] = 'green'
color_map[Flatten]['fill'] = 'teal'

font = ImageFont.truetype("arial.ttf", 32)

visualkeras.layered_view(model, type_ignore=[visualkeras.SpacingDummyLayer])
visualkeras.layered_view(model, type_ignore=[visualkeras.SpacingDummyLayer],
                         legend=True, font=font)
#visualkeras.layered_view(model, spacing=0)
#visualkeras.layered_view(model,type_ignore=[ZeroPadding2D, Dropout, Flatten, visualkeras.SpacingDummyLayer])
#visualkeras.layered_view(model,color_map=color_map, type_ignore=[visualkeras.SpacingDummyLayer])

```
